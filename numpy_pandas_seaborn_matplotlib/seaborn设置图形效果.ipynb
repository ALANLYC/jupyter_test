```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import Series,DataFrame
```

# axes_style and set_style

```python
%matplotlib inline
```

```python
x = np.linspace(0,14,100)
```

```python

y1 = np.sin(x)
y2 = np.sin(2*x)+1.25
```

```python
def sinplot():
    plt.plot(x,y1)
    plt.plot(x,y2)
```

```python
sinplot()
```

```python
style = ['darkgrid','dark','white','whitegrid','tricks']
```

```python
sns.set_style(style[3])
```

```python
sinplot()
```

```python
sns.axes_style()
```

```python
sns.set_style(style[0],{'grid.color':'red'})
```

```python
sinplot()
```

```python
sns.axes_style()
```

```python
sns.set()  #
```

```python
sinplot()
```
