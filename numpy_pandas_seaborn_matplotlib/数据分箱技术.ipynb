```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
#Binning技术
```

```python
score_list = np.random.randint(40,100,size=20)
```

```python
score_list
```

```python
bins = [0,59,70,80,100] #定义取值范围
```

```python
score_cat = pd.cut(score_list,bins)
```

```python
pd.value_counts(score_cat)  #统计各个段的人数
```

```python
df = DataFrame()
```

```python
df['score'] = score_list
```

```python
df["student"] = [pd.util.testing.rands(3) for i in range(20)]  #生成随机字符串
```

```python
df['Categories']=pd.cut(df['score'],bins,labels=['Low','OK','Good','Great'])
```
