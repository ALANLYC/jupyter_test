```python
b = np.arange(0.0,2.0,0.1)
b.size
```

```python
s = np.sin(b*np.pi)
```

```python
plt.plot(b,s,'b--',label='aa')
plt.plot(b*2,s,'y*',label ='bb')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin')
plt.legend()
```

# subplot

```python
plt.subplot(2,1,1)
plt.plot(b,s,'b--',label='aa')
plt.subplot(2,1,2)
plt.plot(b*2,s,'y*',label ='bb')
```

# subplots

```python
figure,ax = plt.subplots(2,2)
ax[0][0].plot(b,s,'b--',label='aa')
ax[0][1].plot(b*2,s,'y*',label ='bb')
plt.show()
```
