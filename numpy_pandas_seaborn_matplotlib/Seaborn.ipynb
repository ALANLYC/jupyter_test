# 热力图

```python
df = sns.load_dataset('flights')
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.pivot(index='month',columns='year',values='passengers') #生成透视表
```

```python
df
```

```python
df.to_csv('flights')
```

```python
sns.heatmap(df)  #生成热力图
```

```python
sns.heatmap(df,annot=True)
```

```python
sns.heatmap(df,annot=True,fmt='d')
```
