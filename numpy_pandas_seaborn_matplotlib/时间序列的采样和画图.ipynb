```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
t_range = pd.date_range('2019-01-01','2019-12-31')
```

```python
t_range
```

```python
s1 = Series(np.random.rand(len(t_range)),index=t_range)
```

```python
s1
```

```python
#采样
s1['2019-01'].mean() #平均值
```

```python
s1_mouth= s1.resample('M').mean()
```

```python
s1_mouth
```

```python
s1.resample('H').bfill()  #向前填充
```

# 画图

```python
t_range = pd.date_range('2019-01-01','2019-12-31',freq='H')
```

```python
t_range
```

```python
stock_df = DataFrame(index=t_range)
```

```python
stock_df['BABA'] = np.random.randint(80,160,size=len(t_range))
```

```python
stock_df
```

```python
stock_df['TEN'] = np.random.randint(80,160,size=len(t_range))
```

```python
stock_df.head()
```

```python
stock_df.plot()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.show(stock_df.plot())
```

```python
weekly_df = DataFrame()
```

```python
weekly_df['BABA'] = stock_df["BABA"].resample('W').mean()
```

```python
weekly_df['TEN'] = stock_df["TEN"].resample('W').mean()
```

```python
weekly_df
```

```python
weekly_df.plot()
```
